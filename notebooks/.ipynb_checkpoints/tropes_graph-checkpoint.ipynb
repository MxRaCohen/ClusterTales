```python
edge_list = edge_list[['_start', '_end']]
edge_list.head()
```

```python
node_list = node_list[['_id', 'name', 'url']]
node_list.head()
```

```python
# Let's get our graph from trope to trope instead of trope id's 
id_to_trope = dict()
for dict_ in node_list.to_dict(orient='records'):
    id_to_trope[dict_['_id']] = dict_['name']
```

```python
named_edge_list = list()
for edge in list(edge_list.itertuples(index=False)):
    start = edge[0]
    end = edge[1]
    named_edge_list.append((id_to_trope[start], id_to_trope[end]))
```

## Build the Graph

```python
trope_graph = nx.MultiDiGraph()
# Add our edges
keys = trope_graph.add_edges_from(named_edge_list)
```

## Build First Pass Static Recommendations
Given a tv/film, find its tropes, return other tv/films with those tropes

```python
def clean_title(title: str) -> str:
    return ''.join(title.split()).lower()
```

```python
def media_to_tropes(media: str, media_tropes=media_tropes):
    # Transform inputed media to a CleanTitle
    media_clean = clean_title(media)
    media_rows = media_tropes.loc[media_tropes['CleanTitle'] == media_clean]
    return media_rows['Trope']
```

```python
def trope_to_media(trope: str, media_tropes=media_tropes):
    trope_rows = media_tropes.loc[media_tropes['Trope'] == trope]
    return trope_rows['CleanTitle']
```

```python
def media_trope_media_prime(media: str):
    tropes = media_to_tropes(media)
    media_list = []
    for trope in tropes:
        media_list.extend(list(trope_to_media(trope)))
    media_list = list(set(media_list))
    return media_list
```

# Get possible recommendation candidates given input
Leveraging our trope graph, let's draft the candidate generation stage of our recommendation system.

```python
# If the user supplies a trope requirement, verify content satisfies the requirement
def check_required_tropes(content_id, mandatory_tropes):
    present_tropes = set(access.get_tropes_from_media(content_id))
    present_and_required = present_tropes.union(set(mandatory_tropes))
    if len(present_and_required) == len(present_tropes):
        return True
    return False
```

```python
# Input is a list of names of content as selected in the multi-select and the list of 'loved' tropes
def find_candidates(input_list, mandatory_tropes = None):
    original_ids = list()
    for content in input_list:
        original_ids.append(access.get_id_from_title(content))
    
    original_tropes = set()
    for id_ in original_ids:
        for trope in access.get_tropes_from_media(id_):
            original_tropes.add(trope)
    
    original_candidates = set()
    for trope in original_tropes:
        for content in access.get_media_from_tropes(trope):
            if mandatory_tropes:
                if check_required_tropes(content, mandatory_tropes):
                    original_candidates.add(content)
            else:
                original_candidates.add(content)

    # Early return if enough candidates were generated before looking at related tropes
    if len(original_candidates) > 100:
        return (original_ids, original_candidates)
    
    expanded_tropes = set()
    for trope in original_tropes:
        try:
            for neighbor in trope_graph.neighbors(trope):
                expanded_tropes.add(neighbor)
        except:
            pass
    
    candidates = set()
    for trope in expanded_tropes:
        for content in access.get_media_from_tropes(trope):
            if mandatory_tropes:
                if check_required_tropes(content, mandatory_tropes):
                    candidates.add(content)
            else:
                candidates.add(content)
    
    # Return original ids and candidate ids
    return (original_ids, candidates)
```

```python
# Find subset of expanded tags based on tropes
def find_candidate_expanded_tags(candidates):
    starting_df = expanded_tags.loc[candidates]
    bool_cols = ~starting_df.any(bool_only=True)
    only_falses = bool_cols[bool_cols]
    # Get the names of these columns
    cols_only_falses = only_falses.index.to_list()
    # Remove these columns
    df = starting_df.drop(cols_only_falses, axis=1)
    return df
```

```python
# Test this out
original_ids, candidates = find_candidates(['The Wiggles Movie (1997)'], ['Trust Password'])
the_wiggles = find_candidate_expanded_tags(candidates)
```

# Corpus Visualization
Now that we have vectorized versions of our contents in trope-space, let's visualize them.

```python
tsne = TSNE(n_components = 2, metric='manhattan', init='pca', verbose=1)

tsne_data = tsne.fit_transform(expanded_tags)
```

```python
plt.figure(figsize=(16,12))
plt.scatter(tsne_data[:,0],tsne_data[:,1],s=1)
plt.title("tsne in 2-dims on all Content-Trope data")
plt.show()
```

# Corpus Clustering
Now that we have a visualization for our corpus, let's visualize clustering over the corpus.

## Kmeans

```python
# Initialize Clustering
kmeans = KMeans(n_clusters=100)

# Fit
kmeans.fit(expanded_tags)

# Predict
cluster_labels = kmeans.predict(expanded_tags)

# Centers
centers = kmeans.cluster_centers_
```

```python
centers.shape
```

```python
plt.figure(figsize=(16,12))
plt.scatter(tsne_data[:,0],tsne_data[:,1],c=cluster_labels, s=1)
plt.title("Clusters in 2-dims on all Content-Trope data")
plt.show()
```

Almost everything appears to be mapped into a tiny fraction of the overall number of clusters.

# Try Clustering over T-SNE

```python
# Initialize Clustering
kmeans_tsne = KMeans(n_clusters=100)

# Fit
kmeans_tsne.fit(tsne_data)

# Predict
cluster_labels_tsne = kmeans_tsne.predict(tsne_data)

# Centers
centers_tsne = kmeans_tsne.cluster_centers_
```

```python
centers_tsne.shape
```

```python
plt.figure(figsize=(16,12))
plt.scatter(tsne_data[:,0],tsne_data[:,1],c=cluster_labels_tsne, s=1)
plt.scatter(centers_tsne[:,0],centers_tsne[:,1],color='red', s=5)
plt.title("Clustered T-SNE in 2-dims on all Content-Trope data")
plt.show()
```

This looks much better but unfortunately is reproducible on new data as T-SNE is not a transformation that can be applied to new data.

# Compare the K-Means Clusters and the TSNE Clusters
What do the numbers say regarding these approaches?

```python
cluster_dict = dict()
cluster_dict['k_means'] = cluster_labels
cluster_dict['tsne'] = cluster_labels_tsne
cluster_dict['titles'] = [access.get_media_name(x) for x in expanded_tags.index]
cluster_df = pd.DataFrame(cluster_dict, index=expanded_tags.index)
```

```python
cluster_df.k_means.value_counts()
```

Yep, k-means is all lumped together into only 2 of the 100 clusters.

```python
cluster_df.tsne.value_counts()
```

I like the distribution of the tsne-transformed clusters better since they're more evenly distributed. Let's try training a PCA reduction instead and clustering over that to see how it goes. 

# PCA Decomp, Clustering, and Visualization

```python
# Inialize PCA - 
pca = PCA(n_components = .95)

# Fit the data
pca = pca.fit(expanded_tags)

# Transform the data
pca_data = pca.transform(expanded_tags)
```

```python
# Initialize Clustering
kmeans_pca = KMeans(n_clusters=100)

# Fit
kmeans_pca.fit(pca_data)

# Predict
cluster_labels_pca = kmeans_pca.predict(pca_data)

# Centers
centers_pca = kmeans_pca.cluster_centers_
```

```python
plt.figure(figsize=(16,12))
plt.scatter(tsne_data[:,0],tsne_data[:,1],c=cluster_labels_pca, s=1)
plt.title("Clustered PCA in 2-dims on all Content-Trope data")
plt.show()
```
